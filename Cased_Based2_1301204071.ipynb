KMeans implementation

```python
#fungsi ini mementukan random centroid pertama 
def random_centroids(data, k):
  centroids = []
  for i in range(k):
    # menrapkan fungsi ke setiap multiple columns using lambda dan apply
    centr = data.apply(lambda x:float(x.sample())) # menrapkan fungsi ke setiap multiple columns using lambda dan apply
    centroids.append(centr) 
    #concat untuk menambah atau di gabung atau append dataframe
    random = pd.concat(centroids,axis = 1) 
  return random

  #fungsi ini untuk mendapatkan labels dari masing masing kolom terhadap centroid
def get_labels(data, centroids):
  distance = centroids.apply(lambda x: np.sqrt(((data - x) ** 2).sum(axis=1)))
  idx = distance.idxmin(axis=1) 
  return idx


  #fungsi ini membuat centroid baru dengan menghitung rata rata dari label dengan centroids yang awal atau lama
def new_centroids(data,labels, k):
  newMean = data.groupby(labels).apply(lambda x : np.exp(np.log(x).mean())).transpose()
  return newMean



```

```python
#fungsi ini hanya untuk menampilkan plot 
from sklearn.decomposition import PCA
from IPython.display import clear_output

k = 3

def plot_clusters(data,labels,centroids, iteration):
    pca = PCA(n_components=2)

    data2 = pca.fit_transform(data)
    centroids2 = pca.transform(centroids.transpose())

    clear_output(wait=True)

    plt.title('Iteration %s.'%(iteration))
    plt.scatter(x=data2[:,0], y=data2[:,1], c = labels)
    plt.scatter(x=centroids2[:,0], y=centroids2[:,1])
    plt.show()


#disini adalah menampilkan iterasi atau looping sampai tiap centroid tidak mengalami perubahan
max_iteration = 100

centroids = random_centroids(data, k)
old_centroids = pd.DataFrame()
iteration = 1


while iteration < max_iteration and not centroids.equals(old_centroids): # akan berhenti jika tidak sesuai kondisi serta centroid tidak berubah 
  old_centroids = centroids

  labels = get_labels(data, centroids)
  
  centroids = new_centroids(data, labels, k)
  iteration += 1

plot_clusters(data, labels, centroids, iteration)






```
